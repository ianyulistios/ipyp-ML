```python
import numpy as np
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataRaw = pd.read_csv('mushrooms.csv')
data = dataRaw.drop(columns=['cap-surface','gill-attachment','stalk-root','veil-type','veil-color'], axis=1)
data
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

for feature in data.columns:
    print(feature, ':', data[feature].unique())
```

```python
lab_enc = preprocessing.LabelEncoder()

Y = data['class'].values
Y_encode = lab_enc.fit_transform(Y)
Y_inverse = lab_enc.inverse_transform(Y_encode)

for feature in data.columns[1:]:
    data[feature] = lab_enc.fit_transform(data[feature])
    
X = data.drop(columns='class', axis=1).values

print(X)
print(Y_encode)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y_encode, test_size=0.2, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=12, algorithm='auto')
knn.fit(X_train,y_train)

print('Training Score: %.2f%%' % (knn.score(X_train, y_train) * 100))
print('Test Score: %.2f%%' % (knn.score(X_test, y_test) * 100))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=500, n_jobs=-1)
rfc.fit(X_train, y_train)
print('Training Score: %.2f%%' % (rfc.score(X_train, y_train) * 100))
print('Test Score: %.2f%%' % (rfc.score(X_test, y_test) * 100))
```

```python
importances = rfc.feature_importances_
features = data.columns[1:]
sort_indices = np.argsort(importances)[::-1]
sorted_features = []
for idx in sort_indices:
    sorted_features.append(features[idx])
plt.bar(range(len(importances)), importances[sort_indices], align='center');
plt.xticks(range(len(importances)), sorted_features, rotation='vertical');
plt.xlim([-1, len(importances)])
plt.grid(False)
plt.show()
```

```python
top4 = sorted_features[:4]
topX = data.iloc[:,[4,6,7,15]]
print(top4)
print(topX)
print(Y_encode)

X_train, X_test, y_train, y_test = train_test_split(topX, Y_encode, test_size=0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=500, n_jobs=2)
rfc.fit(X_train, y_train)
print('Training Score: %.2f%%' % (rfc.score(X_train, y_train) * 100))
print('Test Score: %.2f%%' % (rfc.score(X_test, y_test) * 100))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=20, algorithm='auto')
knn.fit(X_train,y_train)

print('Training Score: %.2f%%' % (knn.score(X_train, y_train) * 100))
print('Test Score: %.2f%%' % (knn.score(X_test, y_test) * 100))
```

```python
#Model Accuration
y_pred=knn.predict(X_test)
y_pred
print("Akurasi Model" , format(np.mean(y_pred==y_test)))
```
